```python
# Part 1

import numpy as np

# reading the given file and storing respective data

time, rad_vel , error_rad_vel , ignore = np.loadtxt("data/UID_0113357_RVC_001.tbl" , unpack='True')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#plt.plot(time, rad_vel, '.')
plt.errorbar(time, rad_vel, yerr=error_rad_vel, fmt='b.', capsize=2)
plt.xlabel("time [days]")
plt.ylabel("Radial velocity [ms$^{-1}$]")
```

```python
# periodogram algorithm

# period array
period = np.logspace(0, 2 , 10000)

string_length = np.zeros(period.shape[0])
#string_length=np.zeros(10000)


for i in range(period.shape[0]):

    phase = (time % period[i]) / period[i]
    
    # indices of the sorted phase
    index = phase.argsort()
    
    # sorted phase based on the sorted phase indices
    sorted_phase = phase[index]
        
    # sorted radial velocity based on the sorted phase indices    
    rad_vel_sorted = rad_vel[index]
    
    # length difference between each data point
    y_diff = rad_vel_sorted[1:] - rad_vel_sorted[:-1]
    x_diff = sorted_phase[1:] - sorted_phase[:-1]
    pieces_length =  np.sqrt( (x_diff ** 2) + (y_diff ** 2) ) 
    string_length[i] = np.sum(pieces_length)
    
    
'''
NDP = time.shape[0]

for i in range(period.shape[0]):
    #generate phases and store phases in order
    
    #phase = np.zeros(NDP)
    phase = ( time % period[i]) / period[i]

    # find out the order of the phase elemnts   
    ophase=phase.argsort()

    # sort phase by value sorted phase=sphase
    
    # sphase = np.zeros(NDP)
    sphase = phase[ophase]

    #sort velocity by phase in sorted velocity=sradv
    sradv = rad_vel[ophase]
 
    #evaluate line length
    string_lengthlength[p] = sum(np.sqrt(np.diff(sphase)**2+np.diff(sradv)**2))

'''    
    
plt.figure(figsize =(12,8))    
plt.loglog(period, string_length)
plt.ylabel("String length")
plt.xlabel("Period [days]")
plt.ylim(1000,100000)


'''
from matplotlib.ticker import ScalarFormatter
plt.gca().xaxis.set_major_formatter(ScalarFormatter(useOffset=False))
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useOffset=False))
'''
```

```python
'''
print np.where(string_length == string_length.min())
'''

# from the above plot it is seen that the first clear minima lies between period values of 4 and 5 days

# we thus try to find the minima in only that range of teh period - values outside the interval are unimportant
string = string_length.copy()

string[(period<4) | (period>5)] = 99999

min_index = np.argmin(string)

T = period[min_index]
print "Period of first clear minima of string length, T =", T, "days"
```

```python
min_phase = (time % period[min_index]) / period[min_index]

# sorting the minimal phase and sorting radial velocity and its error accordingly
index = min_phase.argsort()
sorted_phase = min_phase[index]
rad_vel_sorted = rad_vel[index]
error_rad_vel_sorted = error_rad_vel[index]

plt.figure(figsize =(12,8)) 
plt.errorbar(sorted_phase,rad_vel_sorted,yerr=error_rad_vel_sorted, fmt = '.', capsize=1)
plt.xlabel("Phase of minimum string length")
plt.ylabel("Radial velocity [ms$^{-1}$]")
```

```python
# curve fit

from scipy.optimize import curve_fit

from math import pi

def fit_fn(x, a, b):
    return a * np.sin(2*pi*x + b)

popt, pcov = curve_fit(fit_fn, sorted_phase, rad_vel_sorted)

# as the variance of the parameters a and b lies on the diagonal of the covariance matrix
perr = np.sqrt(np.diag(pcov))

# amplitude of velocity
K = np.abs(popt[0])
K_err = perr[0]

# parameter a implies amplitude of velocity, here K
print "a (here K)= (", K, "+/-", K_err , ") m/s" 
print "b =", popt[1], "+/-", perr[1]

# semi-amplitude
semi_a = K/2
semi_a_err = perr[0]/2

plt.figure(figsize = (12,8))
plt.errorbar(sorted_phase,rad_vel_sorted,yerr=error_rad_vel_sorted, fmt = '.', capsize=1)
plt.plot(sorted_phase, fit_fn(sorted_phase, popt[0], popt[1]), 'r-')
plt.xlabel("Phase of minimum string length")
plt.ylabel("Radial velocity [ms$^{-1}$]")
```
