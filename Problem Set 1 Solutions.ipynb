```python
# Part 1

# Question 1

def reverse_complement(DNA):
    str = ""
    for i in range(len(DNA)-1,-1,-1):
        if DNA[i] == 'A':
            str = str + 'T'
        elif DNA[i] == 'T':
            str = str + 'A'
        elif DNA[i] == 'C':
            str = str + 'G'
        elif DNA[i] == 'G':
            str = str + 'C'
    return str

a = "ACTTGCTATCGTAGCATGGGCGTATTATTAGGCTAGTCGGATTGCGATCGGATCGATTGCA"

print(reverse_complement(a))
    
    
```

```python
# Question 4

# dictionary relating codons to corresponding amino acids
codons = {}
f1 = open('data/problem_1_codons.txt', 'r')
for line in f1:
    line = line.strip()
    columns = line.split()
    codons[columns[0]] = columns [1]
    
# converting dna sequence to mrna
def dna_to_mrna(DNA):
    str = ""
    for i in range(len(DNA)):
        if DNA[i] == 'T':
            str = str + 'U'
        else:
            str = str + DNA[i]
    return str

# converting mrna sequence - each codon - 3 nucleotides into protein (amino acids)
def mrna_to_protein(mrna):
    amino = ""
    for i in range(0,len(mrna),3):
        seq = mrna[i:i+3]
        if codons[seq] == "Stop":
            break
        else:
            amino = amino + codons[seq]
    return amino

# converting a dna sequence into mrna and then into protein
def dna_to_protein(dna):
    mrna = dna_to_mrna(dna)
    protein = mrna_to_protein(mrna)
    return protein

# reading from FASTA file and making dictionary
def read_fasta(filename):
    e = {}
    f2 = open(filename, 'r')
    label = ""
    seq = ""
    for line in f2:
        l = line.strip()
        if l[0] == '>':
            e[label] = seq
            label = l[1:]
            seq = ""
        else:
            seq = seq + l
    
    try:
        del e[""]
    except KeyError:
        pass
    
    return e

filename = 'data/problem_1_question_4.fasta'
dna_dict = read_fasta(filename)

for key in dna_dict:
    print key, ":" , dna_to_protein(dna_dict[key])
        
```

```python
# Part 3

# Question 5

# reading from FASTA file and making dictionary
def read_fasta(filename):
    e = {}
    f2 = open(filename, 'r')
    label = ""
    seq = ""
    for line in f2:
        l = line.strip()
        if l[0] == '>':
            e[label] = seq
            label = l[1:]
            seq = ""
        else:
            seq = seq + l
    
    try:
        del e[""]
    except KeyError:
        pass
    
    return e

# finding the consensus sequence
def consensus_sequence(dna_dict):
    
    length = 0
    
    for key in dna_dict:
        seq = dna_dict[key]
        length = len(seq)
        break
    
    consensus = ""
    index_label = {0 : 'A' , 1 : 'T' , 2 : 'C' , 3 : 'G'}
    
    for i in range(length):
        count = [0,0,0,0]
        # indices 0,1,2,3 correspond to A,T,C and G respectively
        for key in dna_dict:
            seq = dna_dict[key]
            if seq[i] == 'A':
                count[0] = count[0] + 1
            elif seq[i] == 'T':
                count[1] = count[1] + 1
            elif seq[i] == 'C':
                count[2] = count[2] + 1
            elif seq[i] == 'G':
                count[3] = count[3] + 1
                
        max_index = count.index(max(count))
        consensus = consensus + index_label[max_index]
        
    return consensus
        
filename = 'data/problem_1_question_6.fasta'
dna_dict = read_fasta(filename)

print consensus_sequence(dna_dict)
```
